```python
import numpy as np
import matplotlib.pyplot as plt
pi=np.pi
pi
```

```python
y=(np.linspace(0,20,100))
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
x=np.cos(y)
plt.plot(y,x,'r-')
plt.xlabel("Angle")
plt.ylabel("cosine")
plt.title("cosine ")
plt.show()
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)
plt.subplot(2, 1, 1)
plt.plot(x1,y1,'r-')
plt.xlabel("Angle")
plt.ylabel("cosine")
plt.title("damped")


plt.subplot(2, 1, 2)
plt.plot(x2,y2,'b-')
plt.xlabel("Angle")
plt.ylabel("cosine")
plt.title("undamped")
plt.show()





```
